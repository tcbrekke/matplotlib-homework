```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import os
```

```python
clinical_df = pd.read_csv(os.path.join('raw_data','clinicaltrial_data.csv'))
mouse_df = pd.read_csv(os.path.join('raw_data','mouse_drug_data.csv'))

capo_color = 'red'
keta_color = 'green'
infu_color = 'blue'
plac_color = 'black'
```
